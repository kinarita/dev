# ライブラリ

```python
import pandas as pd
from sklearn.cluster import KMeans # sklearnのcluster機能郡からKMeansを使う
from sklearn.preprocessing import MinMaxScaler # sklearnのpreprocessing機能郡からMixMaxScaler（正規化）を使う
import pickle # Python使用の情報を保持するライブラリ
```

```python
# 初期定義
```

# 初期定義

```python
data_dir = '../data'
input_dir = f'{data_dir}/input'
output_dir = f'{data_dir}/output'
input_file = f'{output_dir}/order_processed.csv'
output_file = f'{output_dir}/order_processed_fix.csv'
model_file = f'{output_dir}/model'
```

# データ読み込み

```python
df = pd.read_csv(input_file)
```

```python
display(df.head())
```

# 学習用にデータを加工

# データを機会学習の粒度に揃える

```python
use_cols = ['顧客ID','売上','カテゴリ']
```

```python
tg_df = df[use_cols]
```

```python
display(tg_df)
```

```python
category = tg_df['カテゴリ'].unique()
```

```python
print(category)
```

```python
tg_df_yoko = []
```

```python
for tg_category in category:
    tmp_df = tg_df.loc[tg_df['カテゴリ']==tg_category].copy()
    tmp_df[f'購買回数_{tg_category}'] = 1
    tmp_df.rename(columns={'売上':f'売上_{tg_category}'}, inplace=True)
    tg_df_yoko.append(tmp_df)
tg_df_yoko = pd.concat(tg_df_yoko, ignore_index=False)
```

```python
display(tg_df_yoko)
```

```python
X_df = tg_df_yoko.groupby("顧客ID").sum()
```

```python
display(X_df)
```

```python
print(X_df.isnull().sum()) # 欠損値の確認
```

# 変数を整える
①カテゴリカル変数の数値化
②欠損値への対応
③外れ値への対処
④スケーリング

```python
scaler = MinMaxScaler() # スケーリングの実施、正規化の処理
```

```python
X_df_scaled = scaler.fit_transform(X_df)
```

```python
print(X_df_scaled)
```

# モデルの学習

```python
model = KMeans(n_clusters=4, random_state=0).fit(X_df_scaled) #KMeans()で定義、fit(X_df_scaled)でデータを渡す　# グループ数は４，乱数のランダム性は０で固定している
```

```python
pred = model.predict(X_df_scaled) # predict()の引数に予測対象データを指定する
```

```python
print(pred)
```

# モデルを評価する

```python
X_df['group'] = pred
```

```python
display(X_df.head())
```

```python
print(X_df["group"].value_counts()) # 件数のカウントはvalue_counts()
```

90％が一つのグループになる場合は、偏りがあるので、外れ値の置換除外をやてみる

# モデルの解釈
各グループにどのような特徴があるのか知ること。てっとり早いのは、それぞれ変数の平均値をグループごとに算出して、特徴を掴むこと。

```python
group_df = X_df.groupby("group").mean()
```

```python
display(group_df)
```

```python
group_rank_df = group_df.rank(ascending=False) # ascending=Falseは降順にするため
```

```python
display(group_rank_df)
```

# データを出力する

```python
X_df.to_csv(output_file) #予測結果の出力
```

```python
pickle.dump(model, open(model_file, 'wb')) #modelは書き込みバイナリモードwbを使う。読む場合はrb pickle.load(open(model_file, 'rb)) 
```
